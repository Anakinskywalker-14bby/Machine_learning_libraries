# Matplotlib


```python
#importing matplotlib lib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

# Ploting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding titles x-axis, y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#ploting with a different color and symbol
plt.plot(x,y,'r.')
plt.show()
```

```python
# plot multiple datasets on one graph
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r*')
plt.plot(x,np.cos(x),'b.')
plt.show()
```

# Bar graph

```python
 fig = plt.figure()
 ax = fig.add_axes([0,0,1,1])
 languages = ['English', 'French', 'Spanish', 'Latin', 'German']
 people = [100, 50, 200, 33, 79]
 ax.bar(languages, people)
 plt.xlabel('LANGUAGES')
 plt.ylabel('NUMBER OF PEOPLE')
 plt.show()
```

# Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 200, 33, 79]
ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

# Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color='b')
plt.show()
```
